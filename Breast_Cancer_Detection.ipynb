## Splitting the Dataset: Train & Test Set

```python
# feature matrix/ Ind. variables
x = dataset.iloc[:, 1:-1].values
# removing the first column i.e 'id' because they are just ids nothing else
# and the last column ie. 'diagnosis_M' as it's our dependent var.
```

```python
x.shape
```

```python
# target variable/ dependent var.
y = dataset.iloc[:, -1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Feature Scaling

```python
# We are applying feature scaling because we want to bring all features to the
# same scale
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
# we apply 'fit_transform()' on train set and only 'transform()' on test test.
# This is done to avoid overfitting.
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2: Model Building

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier_lr = LogisticRegression(random_state=0)
```

```python
classifier_lr.fit(x_train, y_train)
```

```python
y_pred = classifier_lr.predict(x_test)
```

```python
# Checking the performance of the model
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
# Calculating Accuracy, F1 Score, Precision & Recall Score of the Model
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test,y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
```

```python
results = pd.DataFrame([['Logistic Regression', acc, f1, prec, rec]],
                       columns = ['Models', 'Accuracy', 'F1 Score', 'Precision', 'Recall'])
```

```python
results
```

```python
# calc the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator=classifier_lr, X=x_train, y=y_train, cv =10)
```

```python
print('Accuracy is {:.2f} %'.format(accuracies.mean()*100))
print('Standard Deviation is {:.2f} %'.format(accuracies.std()*100))
```

## 2. Random Forest

```python
model_results = pd.DataFrame([['Random Forest', acc, f1, prec, rec]],
                       columns = ['Models', 'Accuracy', 'F1 Score', 'Precision', 'Recall'])
```

# Part 4: Final Model (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
classifier_final = LogisticRegression(C =0.5, penalty='l2', random_state=0, solver='newton-cg')
classifier_final.fit(x_train, y_train)
```

```python
y_pred = classifier_final.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test,y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
model_results_2 = pd.DataFrame([['Final Logistic Regression', acc, f1, prec, rec]],
                       columns = ['Models', 'Accuracy', 'F1 Score', 'Precision', 'Recall'])

# Concatenate the new results with the existing DataFrame
results = pd.concat([results, model_results_2], ignore_index=True)

results
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier_final, X=x_train, y=y_train, cv =10)
print('Accuracy is {:.2f} %'.format(accuracies.mean()*100))
print('Standard Deviation is {:.2f} %'.format(accuracies.std()*100))
```
